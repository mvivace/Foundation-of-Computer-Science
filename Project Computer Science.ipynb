```python
# creiamo un nuovo dataframe con le sole colonne necessarie

tot_most_year = all_merge[['university_name', 'year_cwur','world_rank_cwur', 'year_times','world_rank_times', 'year_shanghai', 'world_rank_shanghai']].copy()
tot_most_year
```

```python
# trasformiamo le date delle tre colonne riferite alle date nel formato corretto con pd.to_date time, modificando solo l'anno perchè mese e giorno non sono presenti

tot_most_year['year_times'] = pd.to_datetime(tot_most_year['year_times'], format = '%Y').copy()
tot_most_year['year_shanghai'] = pd.to_datetime(tot_most_year['year_shanghai'], format = '%Y').copy()
tot_most_year['year_cwur'] = pd.to_datetime(tot_most_year['year_cwur'], format = '%Y').copy()
```

```python
tot_most_year
```

```python
# i valori contenuti nella colonna 'world rank times' che indicano un range (ad es. 401-500) vengono separati e messi in due colonne diverse 'rankMinT', 'rankMaxT

datiSplit = tot_most_year['world_rank_times'].str.split('-', expand=True).copy()
tot_most_year[['rankMinT', 'rankMaxT']] = datiSplit 
 
# i valori contenuti nella colonna 'world_rank_shanghai' che indicano un range (ad es. 401-500) vengono separati e messi in due colonne diverse 'rankMinS', 'rankMaxS'

datiSplit2 = tot_most_year['world_rank_shanghai'].str.split('-', expand=True).copy()
tot_most_year[['rankMinS', 'rankMaxS']] = datiSplit2 
tot_most_year
```

```python
# trasformiamo i dati delle tre colonne rank in float

tot_most_year[['rankMinT', 'rankMaxT', 'rankMinS','rankMaxS']] = tot_most_year[['rankMinT', 'rankMaxT', 'rankMinS','rankMaxS']].astype(float)
```

```python
# per ogni riga troviamo il valore massimo e il valore minimo dei rank e li inseriamo nelle colonne 'max_value' e 'min_value'

tot_most_year['max_value'] = tot_most_year.max(axis=1) 
tot_most_year['min_value'] = tot_most_year.min(axis=1) 
```

```python
tot_most_year
```

```python
# funzione per trovare le differenze massime possibili tra i rank di dataframe differenti 

# il primo if viene utilizzato nel caso in cui 'max_value' e 'min_value' coincidano mentre 
# gli altri valori dei rank siano nulli. In tal caso quel valore rappresenterà il 'max_value' 

# dopodichè vengono calcolate tutte le combinazioni di differenze considerando come 
# dividendo e divisore solo valori dei rank appartenenti a dataframe diversi 

# tale funzione per ogni riga ritornerà la differenza massima tra le 16 calcolate

def comparaDifferenze(riga):
    
    if (riga['max_value'] == riga['min_value']): #per le righe in cui vi sono valori nulli 
        return riga['max_value'] 
    
    a = riga['world_rank_cwur'] - riga['rankMinT'] 
    b = riga['world_rank_cwur'] - riga['rankMaxT']
    c = riga['world_rank_cwur'] - riga['rankMinS']  
    d = riga['world_rank_cwur'] - riga['rankMaxS']
    
    e = riga['rankMinT'] - riga['world_rank_cwur'] 
    f = riga['rankMinT'] - riga['rankMinS']
    g = riga['rankMinT'] - riga['rankMaxS']  

    h = riga['rankMaxT'] - riga['world_rank_cwur'] 
    i = riga['rankMaxT'] - riga['rankMinS']
    j = riga['rankMaxT'] - riga['rankMaxS']
    
    k = riga['rankMinS'] - riga['world_rank_cwur'] 
    l = riga['rankMinS'] - riga['rankMinT']
    m = riga['rankMinS'] - riga['rankMaxT']
    
    n = riga['rankMaxS'] - riga['world_rank_cwur'] 
    o = riga['rankMaxS'] - riga['rankMinT']
    p = riga['rankMaxS'] - riga['rankMaxT']
    
    MaxDif = max(a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p)
    return MaxDif
```

```python
# applichiamo la funzione al dataframe tot_most_year, la differenza massima viene poi inserita nella colonna max difference

tot_most_year['maxdifference'] = tot_most_year.apply(comparaDifferenze, axis = 1)
```

### 10. For each university, find the number of rankings in which they appear (it suffices to appear in one year for each ranking).

```python
# leggiamo i 3 file csv denominati 'shangai', 'times', 'cwur'

shanghai = pd.read_csv('/home/monica/projects/fcs/2020/shanghaiData.csv')
times= pd.read_csv('/home/monica/projects/fcs/2020/timesData.csv')
cwur= pd.read_csv('/home/monica/projects/fcs/2020/cwurData.csv')
```

```python
# raggruppiamo per university name e contiamo quante volte l'università compare in ogni ranking, 
# resettiamo index per avere un dataframe e non una serie

shanghai_rank = shanghai.groupby('university_name')['world_rank'].count().reset_index()
times_rank = times.groupby('university_name')['world_rank'].count().reset_index()
cwur_rank = cwur.groupby('institution')['world_rank'].count().reset_index()
```

```python
shanghai_rank
```

```python
# rinominiamo la colonna di cwur_rank 'institution' in 'university_name'

cwur_rank.rename(columns={'institution': 'university_name'}, inplace=True)
```

```python
# creiamo un nuovo dataframe che contiene per ogni università il numero delle apparizioni nei ranking dei diversi dataframe 

merge_rankings_1 = pd.merge(cwur_rank, shanghai_rank, left_on = 'university_name', right_on = 'university_name' , how= 'left', suffixes= ('_cwur', '_shanghai'))
merge_rankings = pd.merge(times_rank, merge_rankings_1, left_on = 'university_name', right_on = 'university_name', how= 'right')

# rinominiamo le colonne 

merge_rankings.rename(columns={'world_rank': 'appearances_times'}, inplace=True)
merge_rankings.rename(columns={'world_rank_cwur': 'appearances_cwur'}, inplace=True)
merge_rankings.rename(columns={'world_rank_shanghai': 'appearances_shanghai'}, inplace=True)

merge_rankings
```

```python
# creiamo un dizionario sentinella le cui chiavi corrispondono ai nomi delle università 
# ed il cui valore iniziale è pari a 0. 

numero_apparizioni = {}
for university in merge_rankings['university_name']:
    numero_apparizioni[university] = 0

# creiamo una funzione che aumenta il valore del dizionario numero_apparizioni ogni volta in cui una 
# università appare in un ranking

def number_rankings (riga):
    if riga['appearances_times'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    if riga['appearances_cwur'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    if riga['appearances_shanghai'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    return(numero_apparizioni)
```

```python
# applichiamo la funzione al dataframe 'merge_rankings'

merge_rankings.apply(number_rankings, axis=1)
numero_apparizioni
```

```python
# convertiamo le colonne in float

tot_rank_year2['world_rank_shanghai'] = tot_rank_year2['world_rank_shanghai'].astype(float)
tot_rank_year2['world_rank_times'] = tot_rank_year2['world_rank_times'].astype(float)
tot_rank_year2['world_rank_cwur'] = tot_rank_year2['world_rank_cwur'].astype(float)
```

```python
# convertiamo l'anno nel formato datetime

tot_rank_year2['year'] = pd.to_datetime(tot_rank_year2['year'], format= '%Y')
```

```python
# creo un dizionario sentinella e scrivo una funzione per paragonare il ranking di ogni uneversità ogni anno.
# viene creato un dizionario in cui associamo alla chiave l'anno, e il valore del ranking per l'università in cui 
#il valore del ranking è uguale in tutti i dataframe

stessa_posizione= {}

def same_position(riga):
    for anno in tot_rank_year2['year']:
        if (riga['world_rank_shanghai'] == riga['world_rank_times']) & (riga['world_rank_shanghai'] == riga['world_rank_cwur']):
            stessa_posizione[riga['university_name']] = (f"year: { riga['year']}, rank: { riga['world_rank_shanghai']}")
        else:
            pass
        return(stessa_posizione)
```
